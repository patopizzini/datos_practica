## Primer Cuatrimestre de 2019. Examen parcial, primera oportunidad

```python
#Leemos el CSV
transacciones = pd.read_csv('transacciones.csv')
transacciones.head(15)
```

```python
#Leemos el CSV
scraping = pd.read_csv('scraping.csv')
scraping.head(15)
```

```python
promedios_T = transacciones.groupby('vehicle_model_id')['price'].agg({'mean','std'}).reset_index().rename(columns={'mean': 'int_mean_price','std': 'int_std_price'})
```

```python
promedios_S = scraping.groupby('vehicle_model_id')['price'].agg({'mean','std'}).reset_index().rename(columns={'mean': 'ext_mean_price','std': 'ext_std_price'})
```

```python
# Se quieren guardar todos los valores de vehiculos, no solo los la empresa.
merged = pd.merge(promedios_S,promedios_T,how='left',on='vehicle_model_id')
```

```python
merged.head(15)
```

## Segundo Cuatrimestre de 2019. Examen parcial, primera oportunidad

## a) Región con mayor cantidad de activos.

```python
#Leemos el CSV
clientes = pd.read_csv('clientes.csv')
clientes.head(15)
```

```python
#Leemos el CSV
eventos = pd.read_csv('eventos.csv')
eventos.head(15)
```

```python
mes_10 = eventos[(eventos['mes'] == 10) & (eventos['anio'] == 2019)] 
```

```python
mes_10.head(15)
```

```python
merged = pd.merge(mes_10[['id_pager']],clientes[['id_pager','region']],how='left',on='id_pager')
```

```python
merged.head(15)
```

```python
merged.drop_duplicates(inplace=True)
```

```python
merged.groupby('region').count().reset_index().nlargest(1, 'id_pager')
```

## b) porcentaje de pagers activos que solo reciben mensajes de un teléfono.

```python
mes_10.head(15) ## Debería darme un 66% CREO.
```

```python
filtrado = mes_10[['id_pager','numero_origen']]
```

```python
filtrado.drop_duplicates(inplace=True)
```

```python
filtrado.head()
```
