# PARCIAL 3
Un sitio de E books tiene información sobre los reviews que los usuarios hacen de sus libros en un DataFrame con formato (user_id, book_id, rating, timestamp).

Por otro lado tenemos información en otro DataFrame que bajamos de GoodReads: (book_id, book_name, avg_rating). Podemos suponer que los Ids de los libros son compatibles.

Se pide usar Python Pandas para:

Obtener un DataFrame que indique el TOP5 de Ebooks en el sitio de Ebooks. (Para este punto se puede ignorar el segundo DataFrame).

Obtener un DataFrame que indique qué libros tienen una diferencia de rating promedio mayor al 20% entre el sitio de Ebooks y GoodReads.

Resolucion: https://github.com/CrossNox/7506-OD2/blob/master/pandas/2017C1_1_ebooks.ipynb

```python
ebooks = pd.read_csv('data/2017C1_1_ebooks.csv')
goodreads = pd.read_csv('data/2017C1_1_goodreads.csv')
```

```python
ebooks.head(50)
```

```python
goodreads.head(50)
```

```python
ebooks_promedio = ebooks.groupby('book_id')[['rating']].mean().reset_index()
ebooks_promedio.columns = ['book_id', 'avg_rating_ebooks']
ebooks_promedio.nlargest(5, 'avg_rating_ebooks')
```

```python
ebooks_goodreads = ebooks_promedio.merge(goodreads, on='book_id')
ebooks_goodreads.head()
```

```python
ebooks_goodreads = ebooks_goodreads[(1-(ebooks_goodreads['avg_rating_ebooks']/ebooks_goodreads['avg_rating'])).abs() > 0.2]
ebooks_goodreads.head(50)
```

```python
def score(s):
    return s.mean()
ebooks_df_mean = ebooks.groupby('book_id').agg({'rating':score})
df_total = goodreads.join(ebooks_df_mean, on='book_id',how='inner')
df_total[(df_total.rating/df_total.avg_rating - 1.0).abs() > 0.2].book_name.tolist()
```

# PARCIAL 4
El GCPD (Gotham City Police Dept) recolecta la información de casos policiales que acontecen en Ciudad Gótica. Esta información se encuentra guardada en un dataframe con el siguiente formato: (fecha, id_caso, descripcion, estado_caso, categoria, latitud, longitud).

Los posibles estados que puede tener un caso son 1: caso abierto, 2: caso resuelto, 3: cerrado sin resolución.

Las fechas se encuentran en el formato YYYY-MM-DD. Por otro lado el comisionado Gordon guarda un registro detallado sobre en cuáles casos fue activada la batiseñal para pedir ayuda del vigilante, Batman. Esta información se encuentra en un Dataframe con el siguiente formato (id_caso, respuesta), siendo campo respuesta si la señal tuvo una respu esta positiva (1) o negativa (0) de parte de él.

El sector encargado de las estadísticas oficiales del GCPD quiere con esta información analizar las siguientes situaciones:

a) Tasa de resolución de casos de la fuerza policial por categoría de caso (considerando aquellos casos en los que no participó Batman).

b) Tasa de resolución de casos con la ayuda de Batman (considerando que aquellos casos en los que fue llamado con la batiseñal, participó en la resolución).

c) Indicar el mes del año pasado en el que Batman tuvo mayor participación en la investigación de casos.

Resolucion: https://github.com/juanlaura/Datos/blob/master/Pandas/Parcial_1C_2018/Parcial%202018%201C.ipynb

```python
gcpd = pd.read_csv('data/2018C1_GCPD.csv')
gordon = pd.read_csv('data/2018C1_gordon.csv')
```

```python
gcpd['estado_caso'].replace(1,'abierto', inplace=True)
gcpd['estado_caso'].replace(2,'resuelto', inplace=True)
gcpd['estado_caso'].replace(3,'sin_resolucion', inplace=True)
gcpd.head(50)
```

```python
gordon.head(50)
```

```python
casos_con_y_sin_batman = gcpd.merge(gordon, how='left', on='id_caso')[['id_caso', 'estado_caso', 'categoria', 'respuesta']].fillna(0)
sin_batman = casos_con_y_sin_batman[casos_con_y_sin_batman['respuesta']!=1].set_index('id_caso')
sin_batman.head(15)
```

```python
por_categoria = sin_batman.groupby(['categoria', 'estado_caso']).count()
por_categoria.head()
```

```python
df_final = por_categoria.unstack(fill_value=0)
df_final.columns = df_final.columns.droplevel(0)
df_final.head()
```

```python
df_final['tasa_resolucion'] = (df_final['resuelto']/(df_final['abierto']+df_final['resuelto']+df_final['sin_resolucion']))*100
df_final.head()
```

# parte B

Tasa de resolución de casos con la ayuda de Batman (considerando que aquellos casos en los que fue llamado con la batiseñal, participó en la resolución).


```python
con_batman = casos_con_y_sin_batman[casos_con_y_sin_batman['respuesta']==1]
con_batman.head()
```

```python
agr_por_caso = con_batman.groupby(['estado_caso'])['id_caso'].agg(['count']).transpose()
agr_por_caso.head()
```

```python
print('la tasa fue de:', agr_por_caso['resuelto'] / (agr_por_caso['abierto']+agr_por_caso['resuelto']+agr_por_caso['sin_resolucion']))
```

```python
agr_por_caso['tasa'] = agr_por_caso['resuelto'] / (agr_por_caso['abierto']+agr_por_caso['resuelto']+agr_por_caso['sin_resolucion']) *100
agr_por_caso.head()
```

```python
con_batman2 = casos_con_y_sin_batman[casos_con_y_sin_batman['respuesta']==1].set_index('id_caso')
por_categoria2 = con_batman2.groupby(['categoria', 'estado_caso']).count()
df_final2 = por_categoria2.unstack(fill_value=0)
df_final2.columns = df_final2.columns.droplevel(0)
df_final2['tasa_resolucion'] = (df_final2['resuelto']/(df_final2['abierto']+df_final2['resuelto']+df_final2['sin_resolucion']))*100
df_final2.head()
```

# parte C

c) Indicar el mes del año pasado en el que Batman tuvo mayor participación en la investigación de casos.

```python
por_fecha = gcpd.merge(gordon, on='id_caso', how='left')[['fecha', 'respuesta']].fillna(0)
por_fecha_batman = por_fecha[por_fecha['respuesta']!=5]
por_fecha_batman['fecha'] = pd.to_datetime(por_fecha_batman['fecha'])
por_fecha_batman['mes'] = por_fecha_batman['fecha'].dt.month
agrupados = por_fecha_batman.groupby('mes')['respuesta'].agg(['count'])
agrupados.nlargest(1, 'count')
```
